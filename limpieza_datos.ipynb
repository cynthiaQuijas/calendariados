## 1. Importar librerías

```python
import pandas as pd
```

## 2. Cargar datos

Se utiliza la base de datos de ejemplo en SQL de Azure

```python
data = pd.read_csv('data.csv')
data.drop(columns=['Unnamed: 0'], inplace=True)
data
```

## 3. Checar valores nulos

```python
print(data.isnull().sum())
```

Hay valores nulos, pero estos son debido a que no aplican a estas personas. Solo debemos checar Title porque esos si deberían tener dato.

```python
data['Title'].unique()
```

```python
data[data['Title'].isnull()]
```

```python
data['Title'].isnull().sum() / len(data)
```

No hay como llenarlos ya que no sabemos si son Mr., Ms., Sr. o Sra. Solo el 0.8% no tiene Title entonces se decidió borrar esos registros.

```python
data.dropna(subset=['Title'], inplace=True)
data['Title'].isnull().sum()
```

## 4. Feature Selection

Variable NameStyle

```python
data['NameStyle'].unique()
```

La variable NameStyle son todos Falsos por lo que esa variable no nos sirve para el modelo

```python
data.drop(columns=['NameStyle'], inplace=True)
```

## 5. Crear csv

```python
# Se crea un csv para los datos con las variables seleccionadas y la variable objetivo
data.to_csv("data_limpio.csv", index=False)
```
