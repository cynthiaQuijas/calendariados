# Implementacion de API

```python
# Importando los datos de prueba
import pandas as pd

# Datos de prueba para probar el modelo
X_test = pd.read_csv("X_test (1).csv")
y_test = pd.read_csv("y_test (1).csv")

if len(X_test) != len(y_test):
    raise ValueError("X_test y y_test no tienen el mismo numero de filas.")

# Combinar los DataFrames
data_test = pd.concat([X_test, y_test], axis=1)

# Mostrar el DataFrame combinado
display(data_test)


```

```python
features = data_test.drop(columns=["DaysSinceModified"])  # Excluir la columna objetivo
#ids = data_test["CustomerID"]  # Guardar los IDs para despliegue

# Convertir caracteristicas a un diccionario para la API
data_dict = features.to_dict(orient='list')
```

```python
features
```

```python
# Formatear los datos para la API
import json
data_json = json.dumps({"data": [data_dict]})

# Leer la URI 
with open("uri.json", "r") as suri:
    scoring_uri = json.load(suri)["URI"][0]
```

```python
# Enviar los datos a la API
import requests
headers = {"Content-Type": "application/json"}
response = requests.post(scoring_uri, data=data_json, headers=headers)
```

```python
response.status_code  # Verificar el estado de la respuesta
```

```python
# Procesar la respuesta de la API
if response.status_code == 200:
    result = json.loads(response.json())  
    data_test["DaysSinceModified"] = result 
    print(result) 
    output = data_test[["CustomerID", "FirstName", "LastName", "DaysSinceModified"]]  
    display(output)  
else:
    print(f"Error: {response.text}")
```
